```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

# **Preparing the dataset**

```python
data = pd.DataFrame(boston.data,columns = boston.feature_names)
data.head()
```

```python
data['Price'] = boston.target
```

```python
data.head()
```

```python
data.info()
```

```python
# check the missing value

data.isnull().sum()
```

```python
# EDA
```

```python
# Correlation

data.corr()
```

```python
sns.pairplot(data)
```

```python
plt.scatter(data['CRIM'],data['Price'])
plt.xlabel('Crime')
plt.ylabel('Price')
```

```python
plt.scatter(data['RM'],data['Price'])
plt.xlabel('RM')
plt.ylabel('Price')
```

```python
sns.regplot(x='RM', y='Price', data=data)
```

```python
sns.regplot(x='CHAS', y='Price', data=data)
```

```python
sns.regplot(x='PTRATIO', y='Price', data=data)
```

```python
# Split the data

x = data.drop('Price', axis=1)
y = data['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.21,random_state=20)
```

```python
# Standardize the data

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
x_train = ss.fit_transform(x_train)
```

```python
x_test = ss.transform(x_test)
```

# **Model Training**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# Print the coefficients and the entercept

print(model.coef_)
```

```python
print(model.intercept_)
```

```python
# Prediction with test data
```

```python
reg_pred = model.predict(x_test)
```

```python
# scatter plot for prediction
plt.scatter(y_test,reg_pred)
```

```python
# residuals means error
residuals = y_test - reg_pred
residuals
```

```python
# plot the residuals

sns.displot(residuals, kind='kde')
```

```python
# plot for prediction ad residuals

plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
```

#  R square 

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,reg_pred)
score
```
